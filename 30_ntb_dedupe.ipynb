```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import scale, minmax_scale, power_transform
from IPython.core.display import HTML
import plotly.express as px
```

```python
sns.set_theme(
    context='talk',
    style='ticks',
    font_scale=.8,
    rc={
        'figure.figsize': (12,8)
    }
)
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.info()
```

```python
df.head(7)
```

```python
numerical_cols = df.select_dtypes(include='number').columns

df.groupby('sex',observed=False)[numerical_cols].mean().reset_index()
```

```python
df.groupby('sex',observed=False)['tip'].mean().reset_index()
```

```python
df.groupby('sex',observed=False)[['tip','total_bill']].mean().reset_index()
```

```python
df_dedupe = df.groupby('sex',observed=False)[['tip','total_bill']].mean().reset_index()
df_dedupe
```

```python
df_dedupe.to_csv('df_dedupe.csv',index=False)
```
